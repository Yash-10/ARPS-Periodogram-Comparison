```python
import pandas as pd

def process(df):
    df[df.columns[1:]] = df[df.columns[1:]].astype(float)
    for col in df.columns[[2, 3, 5, 7, 8, 10]]:
        df[col] = df[col].astype(float)
        df[col] = df[col].round(decimals=2)

    df = df.rename(columns={"Unnamed: 0": "File name"})

    return df

df = pd.read_csv('real_lc_bls_and_tcf_compare.csv')
process(df)
```

Showing the same results as above, but using Gaussian Processes Regression before BLS. TCF results would **not** change.

```python
df = pd.read_csv('real_lc_bls_and_tcf_compare_with_GPR_for_BLS.csv')
process(df)
```

As we see, the results are mostly unchanged. Only slight changes in the SNR are seen. This could be because the light curves were already detrended using spline fits.
